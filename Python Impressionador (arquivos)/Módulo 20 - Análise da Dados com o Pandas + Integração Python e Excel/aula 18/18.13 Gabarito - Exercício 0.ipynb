### Importação de Módulos e Arquivos

```python
import pandas as pd

funcionarios_df = pd.read_csv('CadastroFuncionarios.csv', sep=';', decimal=',')
clientes_df = pd.read_csv('CadastroClientes.csv', sep=';', decimal=',')
servicos_df = pd.read_excel('BaseServiçosPrestados.xlsx')

#retirar colunas Estado Civil e Cargo da tabela de funcionarios
funcionarios_df = funcionarios_df.drop(['Estado Civil', 'Cargo'], axis=1)

display(funcionarios_df)
display(clientes_df)
display(servicos_df)
```

### 1 - Folha Salarial

```python
funcionarios_df ['Sálario Total'] = funcionarios_df['Salario Base'] + funcionarios_df['Impostos'] + funcionarios_df['Beneficios'] + funcionarios_df['VT'] + funcionarios_df['VR']

display(funcionarios_df)

# ordenar o dataframe
funcionarios_df = funcionarios_df.sort_values('Sálario Total', ascending=False)

folha_total = sum(funcionarios_df['Sálario Total'])
print(f'O total da folha salárial mensal é de R$ {folha_total:,}')
```

(EXTRA) **Soma de salário por area**

```python
salarios_por_area_df = funcionarios_df.groupby('Area').sum()
salarios_por_area_df = salarios_por_area_df.drop(['ID Funcionário','Nome Completo'], axis = 1)

# ordenar tabela
salarios_por_area_df = salarios_por_area_df.sort_values('Sálario Total', ascending = False)

display(salarios_por_area_df)
salarios_por_area_df.plot(kind='bar')

# salarios_por_area_df.dtypes -> verificar os tipos das colunas
```

### 2 - Faturamento da Empresa

```python
faturamentos_df = servicos_df[['ID Cliente', 'Tempo Total de Contrato (Meses)']].merge(clientes_df[['ID Cliente', 'Valor Contrato Mensal']], on='ID Cliente')

faturamentos_df['Faturamento Total'] = faturamentos_df['Tempo Total de Contrato (Meses)'] * faturamentos_df['Valor Contrato Mensal']
display(faturamentos_df)

faturamento_total = sum(faturamentos_df['Faturamento Total'])
print(f'O Faturamento Total da empresa é de R$ {faturamento_total:,}')
```

(EXTRA) **Melhores Clientes**

```python
melhores_clientes_df = faturamentos_df[['ID Cliente', 'Faturamento Total']]
melhores_clientes_df = melhores_clientes_df.sort_values('Faturamento Total', ascending = False)[:5]


melhores_clientes_df = melhores_clientes_df.set_index('ID Cliente')

display(melhores_clientes_df)
melhores_clientes_df.plot(kind='bar', xlabel='ID Cliente', ylabel='Faturamento Total')
```

### 3 - % Funcionários Fecharam Contrato

```python
qtd_funcionario_fecharam_contrato = len(servicos_df['ID Funcionário'].unique())
qtd_funcionario_total = len(funcionarios_df['ID Funcionário'])

print(f'Percentual de funcionarios que fecharam contrato {qtd_funcionario_fecharam_contrato / qtd_funcionario_total:.2%}')
```

### 4 - Qtde Contratos por Área

```python
contratos_area_df = servicos_df[['ID Funcionário']].merge(funcionarios_df[['ID Funcionário', 'Area']], on= 'ID Funcionário')
display(contratos_area_df)

qtd_contratos_area = contratos_area_df['Area'].value_counts()

print(qtd_contratos_area)
qtd_contratos_area.plot(kind='bar')
```

### 5 - Funcionários por Área

```python
funcionarios_area = funcionarios_df['Area'].value_counts()
print(funcionarios_area)
funcionarios_area.plot(kind='bar')
```

### 6 - Ticket Médio Mensal

```python
tiquet_medio = clientes_df['Valor Contrato Mensal'].mean()
print(f'O tiquet médio é de R$ {tiquet_medio:,.2f}')
```
